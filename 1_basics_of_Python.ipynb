# 2. Variables

üìò Definition

A variable is a name given to a memory location used to store data.

üìù Rules for Variables

Must start with a letter or underscore _

Cannot start with a number

No spaces allowed

Case-sensitive

üß™ Code Example

```python
name = "Ali"
age = 20
marks = 85.5

print(name)
print(age)
print(marks)
```

### 3. Data Types
üìò Definition

Data types specify the type of data a variable can store.

üìä Common Data Types

| Data Type | Description       | Example    |
| --------- | ----------------- | ---------- |

### 5. Operators in Python
üìò Definition

Operators are symbols used to perform operations on values and variables.

* 1. ‚ûï Arithmetic Operators

Used for mathematical operations.

| Operator | Meaning        |
| -------- | -------------- |
| +        | Addition       |
| -        | Subtraction    |
| *        | Multiplication |
| /        | Division       |
| %        | Remainder      |


```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
```

* 2. üîç Comparison Operators

Used to compare values.

```python
print(5 > 3)
print(5 < 3)
print(5 == 5)
print(5 != 3)
```

* 3. üîó Logical Operators

Used to combine conditions.

```python
print(True and False)
print(True or False)
print(not True)
```
